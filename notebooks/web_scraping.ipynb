```python
def fetch_website_content(url):
    """
    Fetch and clean text content from a given URL.
    """
    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()  # Check if the request was successful
        soup = BeautifulSoup(response.content, "html.parser")
        text = soup.get_text(separator=" ", strip=True)  # Extract and clean text
        return text
    except Exception as e:
        print(f"Error fetching {url}: {e}")
        return None
```

```python
def tokenize_thai_text(text):
    """
    Tokenize Thai text into words and return as a list.
    """
    if not isinstance(text, str):
        return text  # If the input is not a string, return it as is
    return word_tokenize(text, engine="newmm")
```

```python
df_tokens = pd.read_csv('/content/drive/MyDrive/CPE393/Project/Datasets/thai_tokens.csv')
```

```python
def split_texts_to_tokens(df):
    new_rows = []
    for index, row in df.iterrows():
        if isinstance(row['texts'], str):
            tokens = row['texts'].split(',')
            for token in tokens:
                token = token.strip().replace('[', '').replace(']', '').replace("'", "")
                if token:
                    new_rows.append({'tokens': token})
        else:
            pass
    return pd.DataFrame(new_rows)

df_temp = split_texts_to_tokens(df_thai_texts)
df_temp
```

```python
df_tokens = pd.concat([df_tokens, df_temp], ignore_index=True)
df_tokens
```
