```python
from googlesearch import search
import pandas as pd
```

Web Crawling

```python
urls = []
```

```python
def crawl_websites(keyword, num_new_links):
    """
    Search for websites related to a specific keyword and return exactly `num_new_links` unique links
    that are not already in the global `urls` list.

    Args:
    - keyword (str): The search keyword.
    - num_new_links (int): The number of new links to retrieve.

    Returns:
    - List of new links.
    """
    global urls  # Use the global list to track already seen URLs

    print(f"Searching for websites related to: {keyword}")

    # List to store the new links for this specific search
    new_links = []

    # Continue searching until we gather enough new links
    while len(new_links) < num_new_links:
        try:
            for url in search(keyword, num_results=num_new_links * 2, lang="en"):
                # Check if the URL is already in the global list or new links list
                if url not in urls and url not in new_links:
                    new_links.append(url)
                    print(f"Found new link: {url}")

                # Stop searching if we've reached the required number of new links
                if len(new_links) >= num_new_links:
                    break
        except Exception as e:
            print(f"Error during search: {e}")
            break

    # Add the new links to the global list
    urls.extend(new_links)

    # Print and return the new links
    print("New links = [")
    for link in new_links:
        print(f'    "{link}",')
    print("]")

    return new_links
```
